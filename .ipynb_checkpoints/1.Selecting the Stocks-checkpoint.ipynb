#### Step 3:  Download the data using Yahoo Finance.

```python
sp500 = yf.download(tickers,start=start_date, end=end_date) # Download the data.
```

####  Step 4:  Calculate the Volatility of Stocks.

```python
daily_volatility    = df_close.pct_change().apply(lambda x: np.log(1+x)).std()     # Calculate the daily volatility
weekly_volatility   = daily_volatility.pct_change().apply(lambda x: x*np.sqrt(5))  # Calculate the weekly volatility(considerign 5 days per week.) 
```

```python
WV           =pd.DataFrame(weekly_volatility).reset_index() # Create a new dataframe and reset the index.
WV.columns   =["Tick","Volatility"]                         # Renames the columns to "Tick", "Volatility"
```

```python
sorted_weekly_volatility  = WV.sort_values(by="Volatility",ascending=False) # Sort the values using Volatility in desc.
```

```python
sorted_weekly_volatility.head() # The top 5 or High voaltile stocks.
```

```python
sorted_weekly_volatility.tail(6) # The least volatile stocks.
```
